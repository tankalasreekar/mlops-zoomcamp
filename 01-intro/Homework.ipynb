### Q2. Computing duration

What's the standard deviation of the trips duration in January?

```python
df.head()
```

```python
df.info()
```

```python
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
```

```python
df['duration'] = df['duration'].apply(lambda x: x.total_seconds() / 60)
```

```python
df
```

```python
df.duration.std()
```

### Q3. Dropping outliers

What fraction of the records left after you dropped the outliers?

```python
df = df[(df['duration']>=1) & (df['duration']<=60)]
```

```python
df.shape
```

```python
3009173/3066766
```

### Q4. One-hot encoding

What's the dimensionality of this matrix (number of columns)?

```python
# Feature columns: PULocationID, DOLocationID
# Target column: duration

df_train = df[['PULocationID', 'DOLocationID', 'duration']]
```

```python
df_train.shape
```

```python
df_train[['PULocationID', 'DOLocationID']] = df_train[['PULocationID', 'DOLocationID']].astype(str)
```

```python
df_train.head()
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
# Turn the dataframe into a list of dictionaries
df_train_dict = df_train[['PULocationID', 'DOLocationID']].to_dict(orient = 'records')
```

```python
# Fit a dictionary vectorizer
dv = DictVectorizer()
X_train = dv.fit_transform(df_train_dict)
```

```python
X_train.shape
```

### Q5. Training a model

What's the RMSE on train?

```python
y_train = df_train['duration'].values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
```

```python
root_mean_squared_error(y_train, y_pred)
```

### Q6. Evaluating the model

Now let's apply this model to the validation dataset (February 2023).
What's the RMSE on validation?

```python
df_val = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet')
```

```python
df_val['duration'] = df_val.tpep_dropoff_datetime - df_val.tpep_pickup_datetime
df_val['duration'] = df_val['duration'].apply(lambda x: x.total_seconds() / 60)

df_val = df_val[(df_val['duration']>=1) & (df_val['duration']<=60)]

df_val = df_val[['PULocationID', 'DOLocationID', 'duration']]
df_val[['PULocationID', 'DOLocationID']] = df_val[['PULocationID', 'DOLocationID']].astype(str)

df_val_dict = df_val[['PULocationID', 'DOLocationID']].to_dict(orient = 'records')

X_val = dv.transform(df_val_dict)
y_val = df_val['duration'].values

y_val_pred = lr.predict(X_val)
root_mean_squared_error(y_val, y_val_pred)
```
